```python
'''Q.1) Assign different types of values to different variables, print them and display all the
variables which you have created. And display the types of all variables and delete a
particular variable. Assign two variables with integer values then print and compare both
variables.'''
a = 10
b = 3.14
c = "Hello"
d = True
e = [1, 2, 3]
f = (4, 5, 6)
g = {"name": "Arman"}


print("a =", a, "| type:", type(a))
print("b =", b, "| type:", type(b))
print("c =", c, "| type:", type(c))
print("d =", d, "| type:", type(d))
print("e =", e, "| type:", type(e))
print("f =", f, "| type:", type(f))
print("g =", g, "| type:", type(g))


del e
print("\nAfter deleting variable 'e':")

x = 25
y = 40

print("\nx =", x)
print("y =", y)

if x > y:
    print("x is greater than y")
elif x < y:
    print("y is greater than x")
else:
    print("x and y are equal")
```

```python
'''Q.2) Write Python program to input a 3-digits integer number then find and
display its reverse (using operators only).'''
num=int(input("Enter a 3-digit integer:"))

hundreds = num // 100
tens = (num // 10) % 10
ones = num % 10

reverse_num = ones * 100 + tens * 10 + hundreds

print("Original number:", num)
print("Reversed number:", reverse_num)
```

```python
# Q3) Find biggest of two numbers using operators only

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("The biggest number is:", a)
elif b > a:
    print("The biggest number is:", b)
else:
    print("Both numbers are equal.")
```

```python
# Q4) Initialize a set and check membership using 'in' operator

my_set = {10, 20, 30, 40, 50}

print("Set is:", my_set)
element = int(input("Enter an element to search in the set: "))

if element in my_set:
    print(element, "is present in the set.")
else:
    print(element, "is NOT present in the set.")
```

```python
# Q5) Distance between two points using only operators

x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))
x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

dx = x2 - x1
dy = y2 - y1

distance = (dx**2 + dy**2) ** 0.5

print("Distance between the two points is:", distance)
```

```python
# Q6) Replace an existing substring with a new string

text = input("Enter the original string: ")
old_sub = input("Enter the substring to be replaced: ")
new_sub = input("Enter the new substring: ")

new_text = text.replace(old_sub, new_sub)

print("Original string: ", text)
print("Modified string: ", new_text)
```

```python
# Q8) Show characters from the string as per the given range

s = input("Enter a string: ")
start = int(input("Enter start index (0-based): "))
end = int(input("Enter end index (exclusive): "))

substring = s[start:end]

print("Characters from index", start, "to", end, "are:", substring)
```

```python
# Q9) Display the given string in reverse order

s = input("Enter a string: ")

reverse_s = s[::-1]

print("Original string:", s)
print("Reversed string:", reverse_s)
```

```python
# Q10) Capitalize the first letter of each and every word of a string

s = input("Enter a string: ")
words = s.split()
cap_words = [word.capitalize() for word in words]
result = " ".join(cap_words)

print("Original string:", s)
print("Capitalized string:", result)
```

```python
# Q11) Check if a 3-digit number is palindrome without using any loop

num = int(input("Enter a 3-digit number: "))
hundreds = num // 100
tens = (num // 10) % 10
ones = num % 10

reverse_num = ones * 100 + tens * 10 + hundreds

print("Original number:", num)
print("Reversed number:", reverse_num)

if num == reverse_num:
    print("It is a palindrome number.")
else:
    print("It is NOT a palindrome number.")
```
